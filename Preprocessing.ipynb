```python
import pandas as pd
import numpy as np 
```

```python
# a list with all missing value formats
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]
df = pd.read_csv("risk-train.txt", delim_whitespace=True, na_values = missing_value_formats)
```

```python
# Number of missing values per column
df_missValue_num = df.isna().sum()
df_missValue_num.head(len(df_missValue_num))
```

```python
# Percentage of missing values per column
df_missValue_percnt = df.isna().sum()/len(df)*100
df_missValue_percnt.head(len(df_missValue_percnt))
```

```python
# we keep only columns where there are at least the 80% of not missing values
df.dropna(thresh=0.8*len(df),axis=1,inplace=True)
df.head()
```

```python
# load original data again
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]
df = pd.read_csv("risk-train.txt", delim_whitespace=True, na_values = missing_value_formats)
df.head(10)
```

```python
# status of missing values per column
cols_missingValue_status = df.isna().sum() > 0
print(cols_missingValue_status)
```

```python
# All columns with missing values
cols_with_missingValues = cols_missingValue_status[cols_missingValue_status == True]
print(cols_with_missingValues)
```

```python
# All columns with missing values' percentage above %70
threshold_percnt_missValue = 70.0
columns_with_massive_missingValue = []
for col in df_missValue_percnt.index:
    if df_missValue_percnt[col] >= threshold_percnt_missValue:
        columns_with_massive_missingValue.append(col)
print(columns_with_massive_missingValue)
```

```python
# Drop all columns when their percentage of the missing values is above %70
for colName in columns_with_massive_missingValue:
    df = df.drop([colName],axis=1)
df.head()
```

Columns 'B_BIRTHDATE' and 'TIME_ORDER' 

```python
# We can delete records with missing values in columns 'B_BIRTHDATE' and 'TIME_ORDER' 
# because they do not have a large number of them.
df.dropna(subset=['B_BIRTHDATE','TIME_ORDER'])
```

Column 'Z_LAST_NAME'

```python
# Column 'Z_LAST_NAME' determines that whether is the name of the account or cardholder identical to 
# the name of the delivery address or not (yes/no).
# We can consider 'yes' as a default value here. 
df.Z_LAST_NAME = df.Z_LAST_NAME.fillna('yes')
```

```python
df.Z_LAST_NAME.head(10)
```

Column 'Z_CARD_ART'

```python
# Column 'Z_CARD_ART' is selected type of card if payment by card is chosen.
# Missing values can be replaced by 'Unknown'
df['Z_CARD_ART'] = df['Z_CARD_ART'].replace(np.nan, 'Unknown')
```

```python
df.head()
```

Column 'DATE_LORDER'

```python
# Column 'DATE_LORDER' is Date of the last order, if available
# We can replace missing values with first Date in this column.
min_col_DATE_LORDER = df[df["DATE_LORDER"].notna()].DATE_LORDER.values.min()
print(min_col_DATE_LORDER)
```

```python
df['DATE_LORDER'] = df['DATE_LORDER'].replace(np.nan, min_col_DATE_LORDER)
```

```python
df.head(20)
```

Column 'MAHN_AKT'

```python
# Column 'MAHN_AKT' is Current stage of reminder of the customer, if available
# We can consider a new stage and it could be maximum + 1
max_col_MAHN_AKT = df[df["MAHN_AKT"].notna()].MAHN_AKT.values.max()
print(max_col_MAHN_AKT)
```

```python
max_col_MAHN_AKT += 1
df['MAHN_AKT'] = df['MAHN_AKT'].replace(np.nan, max_col_MAHN_AKT)
```

```python
df.head()
```

Column 'MAHN_HOECHST'

```python
# Column 'MAHN_HOECHST' is Current stage of reminder of the customer, if available
# We can consider a new stage and it could be maximum + 1
max_col_MAHN_HOECHST = df[df["MAHN_HOECHST"].notna()].MAHN_HOECHST.values.max()
print(max_col_MAHN_HOECHST)
```

```python
max_col_MAHN_HOECHST += 1
df['MAHN_HOECHST'] = df['MAHN_HOECHST'].replace(np.nan, max_col_MAHN_HOECHST)
```
